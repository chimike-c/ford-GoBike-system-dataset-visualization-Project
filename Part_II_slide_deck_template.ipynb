## Investigation Overview

For this presentataion, we aim to show case data driven insights obtained from analysis on the Ford GoBike system dataset. 
The following insights have been drawn and are to be analyized; 


-  The bike share for all feature is not available to 'one-time'  customers, only to subscribers. This feature can be optimized and can be leveraged to sell subscription to customers. 

-  The wide spread of user demogarphic (sex and age) should allow the company venture into more sub-services for the various demographics e.g. bike buddy(ride with a friend), or keep fit challenges between friends, etc. 



## Dataset Overview

The insights are drawn from the Ford GoBike system dataset. This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python

df_1 = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
# forming a member age column from the member birth year data
member_age = 2022 - df_1.member_birth_year 

df_1['member_age'] = member_age
```

## (Visualization 1)

This first visualisation shows the trend observed between user_type and bike_share_for_all_trip. 


```python
plt.figure(figsize = [8, 8])
sb.countplot(data = df_1, x = 'user_type', hue = 'bike_share_for_all_trip', palette = 'Greens')

plt.title('Trend between user type and bike share for all trip varibales')
plt.show()
```

### Comments on the visualisation

The above plot shows the trend between user types and bike share for all trip variables. As we can see from the plot, none of the users under the customer user type use the bike share for all trip service feature. 

- The bike share for all feature is not available to 'one-time'  customers, only to subscribers. This feature can be optimized and can be leveraged to sell subscription to customers.

## (Visualization 2)

This next visualisation shows the trend observed for the member age distribution. 

```python
binsize = 1
bins = np.arange(0, df_1['member_age'].max()+binsize, binsize)

plt.figure(figsize=[8, 8])
plt.hist(data = df_1, x = 'member_age', bins = bins)
plt.xlabel('Member age')
plt.ylabel('Count')

plt.title('Range of member age values for GoBike Users')
plt.show()
```

### Comments on the visualisation

The member age distribution shows the wide range of ages of the Ford GoBike users. As a result of this vast range, the compamny can  venture into more sub-services for the various demographics (older users and younger users) e.g. bike buddy(ride with a friend), or keep fit challenges between friends, etc.

- The distribution shows that a larger proportion of users fall under the age range of 20-55 who are more susceptible to such 
interactive programs, hence the suggested sub-services would do well for the company. 

## (Visualization 3)

The next visualisation shows the distribution for member gender. 

```python
plt.figure(figsize = [8,8])

default_color = sb.color_palette()[0]
sb.countplot(data = df_1, x = 'member_gender', color = default_color)

plt.title('Trend showing member gender amongst users')
plt.show()
```

### Comments on the visualisation

The distribution shows a diverse gender range of users, with a large proportion being men and relatively smaller proportion being women and others. This shows that there are way more male users than other users. 

- The company can levergae this to add more diverse marketing campaigns to attract more females and users of other gender categories, and also to keep the number of male users up.
